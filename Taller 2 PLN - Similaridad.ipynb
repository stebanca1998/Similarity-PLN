```python
#Steban Cadena Giraldo
#1670129-3743

import numpy as np
import math
from matplotlib import  pyplot as plt

qv = np.array([13,22])
doc1 = np.array([27,14])
doc2 = np.array([4,0])
doc3 = np.array([24,17])
```

```python
def simcos(q,d):
    productopunto = np.dot(q,d)
    normaQ = np.linalg.norm(q)
    normaD = np.linalg.norm(d)
    
    res = productopunto / (normaQ * normaD)
    
    return res

def sim(q,d):
    productopunto = np.dot(q,d)
    denominador=0
    
    for i in range(len(q)):
        denominador += (q[i]**2)*(d[i]**2)
    
    denominador = math.sqrt(denominador)
    
    res = productopunto/denominador
    
    return res
```

```python
print("Similaridad del coseno")
print("Entre q y doc1 = ", simcos(qv,doc1))
print("Entre q y doc2 = ", simcos(qv,doc2))
print("Entre q y doc3 = ", simcos(qv,doc3) , "\n")

print("Similaridades por sim")
print("Entre q y doc1 = ", sim(qv,doc1))
print("Entre q y doc2 = ", sim(qv,doc2))
print("Entre q y doc3 = ", sim(qv,doc3))
```
